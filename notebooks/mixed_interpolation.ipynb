```python
import rateslib as rl
```

```python
mixed_curve = rl.Curve(
    nodes={
        rl.dt(2022, 1, 1): 1.0,
        rl.dt(2023, 1, 1): 0.99,
        rl.dt(2024, 1, 1): 0.979,
        rl.dt(2025, 1, 1): 0.967,
        rl.dt(2026, 1, 1): 0.956,
        rl.dt(2027, 1, 1): 0.946,
    },
    interpolation="log_linear",
    t=[
        rl.dt(2024, 1, 1),
        rl.dt(2024, 1, 1),
        rl.dt(2024, 1, 1),
        rl.dt(2024, 1, 1),
        rl.dt(2025, 1, 1),
        rl.dt(2026, 1, 1),
        rl.dt(2027, 1, 1),
        rl.dt(2027, 1, 1),
        rl.dt(2027, 1, 1),
        rl.dt(2027, 1, 1),
    ],
)
```

```python
mixed_curve.plot("1D", comparators=[mixed_curve], labels=["log-linear", "log-cubic-mix"])
```
